```python
tags = [dict(tag) for tag in oahu['tags'].to_list() if tag]
```

```python
len(tags)
```

```python
all_tag_types = list(reduce(lambda a, b: a | b, [set(dicts.keys()) for dicts in tags]))
```

```python
len(all_tag_types)
```

```python
for i, tag in enumerate(sorted(all_tag_types)):
    if i%5 == 0:
        print(f'\n{tag}\t', end= " ")
    else:
        print(f'{tag}\t', end= " ")
    
```
